```python
h = f['p8_ee_Zmumu_ecm91/evts_initial;1']

y1,x = h.to_numpy()
plt.plot(x[:-1],y1, color='rosybrown', drawstyle = 'steps-mid')
mumu_initial = sum(y1)
print(mumu_initial)

plt.title("initial events: mumu")
plt.xlim(0,3)
```
